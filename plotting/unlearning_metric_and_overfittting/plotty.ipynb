```python
import numpy as np
import pandas as pd

c1=pd.read_csv('catastrophic_unlearning_stats_epochs1.csv')
c2=pd.read_csv('catastrophic_unlearning_stats_epochs2.csv')
c3=pd.read_csv('catastrophic_unlearning_stats_epochs3.csv')
c4=pd.read_csv('catastrophic_unlearning_stats_epochs4.csv')
c5=pd.read_csv('catastrophic_unlearning_stats_epochs5.csv')


m1=pd.read_csv('modular_unlearning_stats_epochs1.csv')
m2=pd.read_csv('modular_unlearning_stats_epochs2.csv')
m3=pd.read_csv('modular_unlearning_stats_epochs3.csv')
m4=pd.read_csv('modular_unlearning_stats_epochs4.csv')
m5=pd.read_csv('modular_unlearning_stats_epochs5.csv')


r1=pd.read_csv('retraining_unlearning_stats_epochs1.csv')
r2=pd.read_csv('retraining_unlearning_stats_epochs2.csv')
r3=pd.read_csv('retraining_unlearning_stats_epochs3.csv')
r4=pd.read_csv('retraining_unlearning_stats_epochs4.csv')
r5=pd.read_csv('retraining_unlearning_stats_epochs5.csv')



UM_C=[c1['unlearning_metric'],c2['unlearning_metric'],c3['unlearning_metric'],c4['unlearning_metric'],c5['unlearning_metric']]
UM_M=[m1['unlearning_metric'],m2['unlearning_metric'],m3['unlearning_metric'],m4['unlearning_metric'],m5['unlearning_metric']]
UM_R=[r1['unlearning_metric'],r2['unlearning_metric'],r3['unlearning_metric'],r4['unlearning_metric'],r5['unlearning_metric']]


OM_C=[c1['overfitting_metric'],c2['overfitting_metric'],c3['overfitting_metric'],c4['overfitting_metric'],c5['overfitting_metric']]
OM_M=[m1['overfitting_metric'],m2['overfitting_metric'],m3['overfitting_metric'],m4['overfitting_metric'],m5['overfitting_metric']]
OM_R=[r1['overfitting_metric'],r2['overfitting_metric'],r3['overfitting_metric'],r4['overfitting_metric'],r5['overfitting_metric']]

MIA_C=[c1['mia_score'],c2['mia_score'],c3['mia_score'],c4['mia_score'],c5['mia_score']]
MIA_M=[m1['mia_score'],m2['mia_score'],m3['mia_score'],m4['mia_score'],m5['mia_score']]
MIA_R=[r1['mia_score'],r2['mia_score'],r3['mia_score'],r4['mia_score'],r5['mia_score']]

# plot UM_C, UM_M simulataneously
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5),dpi=150)
plt.plot([1,2,3,4,5],UM_C, label='Catrastrophic', marker='o', linestyle='-', color='b')
plt.plot([1,2,3,4,5],UM_M, label='Modular', marker='o', linestyle='-', color='g')
plt.plot([1,2,3,4,5],UM_R, label='Retraining', marker='o', linestyle='-', color='r')
plt.xlabel('Epochs')
plt.ylabel('Unlearning Metric')
plt.xticks([1,2,3,4,5])
plt.legend()
plt.grid()
plt.show()



# plot OM_C, OM_M simulataneously
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5),dpi=150)
plt.plot([1,2,3,4,5],OM_C, label='Catrastrophic', marker='o', linestyle='-', color='b')
plt.plot([1,2,3,4,5],OM_M, label='Modular', marker='o', linestyle='-', color='g')
plt.plot([1,2,3,4,5],OM_R, label='Retraining', marker='o', linestyle='-', color='r')
plt.xlabel('Epochs')
plt.ylabel('Overfitting Metric')
plt.xticks([1,2,3,4,5])
plt.legend()
plt.grid()
plt.show()


# plot MIA_C, MIA_M simulataneously
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5),dpi=150)
plt.plot([1,2,3,4,5],MIA_C, label='Catrastrophic', marker='o', linestyle='-', color='b')
plt.plot([1,2,3,4,5],MIA_M, label='Modular', marker='o', linestyle='-', color='g')
plt.plot([1,2,3,4,5],MIA_R, label='Retraining', marker='o', linestyle='-', color='r')
plt.xlabel('Epochs')
plt.ylabel('MIA Score')
plt.xticks([1,2,3,4,5])
plt.legend()
plt.grid()
plt.show()
```
