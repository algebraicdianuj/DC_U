```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Function to convert to dB
def to_dB(value):
    return 10 * np.log10(value)
# Reading the CSV files
DM = pd.read_csv("benchmark_DM_without_DSA.csv")
GM = pd.read_csv("benchmark_GM_without_DSA.csv")
IDM = pd.read_csv("benchmark_proposed.csv")
MI = pd.read_csv("benchmark_proposed_mi_lite.csv")

bar_width = 0.2
r1 = np.arange(len(DM['IPC']))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]

# Convert running times to dB scale
DM['Running Time dB'] = DM['Running Time'].apply(to_dB)
GM['Running Time dB'] = GM['Running Time'].apply(to_dB)
IDM['Running Time dB'] = IDM['Running Time'].apply(to_dB)
MI['Running Time dB'] = MI['Running Time'].apply(to_dB)
# Choose a color palette
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# Plot for Running Time
plt.figure(figsize=(12, 6), dpi=300)

plt.bar(r1, DM['Running Time dB'], width=bar_width, label='Distribution Matching', alpha=0.8, color=colors[0])
plt.bar(r2, GM['Running Time dB'], width=bar_width, label='Gradient Matching', alpha=0.8, color=colors[1])
plt.bar(r3, IDM['Running Time dB'], width=bar_width, label='Improved Distribution Matching (Proposed)', alpha=0.8, color=colors[2])
plt.bar(r4, MI['Running Time dB'], width=bar_width, label='Model Inversion (Proposed)', alpha=0.8, color=colors[3])

plt.xlabel('IPC', fontsize=14)
plt.ylabel('Running Time (dBs)', fontsize=14)
# plt.title('Comparison of Running Time Across Benchmarks', fontsize=16)
plt.xticks([r + 1.5*bar_width for r in range(len(DM['IPC']))], DM['IPC'], fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)

plt.tight_layout()
plt.savefig("Running_Time_Comparison.png")
plt.show()

# Plot for Test Accuracy
plt.figure(figsize=(12, 6), dpi=300)

plt.bar(r1, DM['Test Accuracy'], width=bar_width, label='Distribution Matching', alpha=0.8, color=colors[0])
plt.bar(r2, GM['Test Accuracy'], width=bar_width, label='Gradient Matching', alpha=0.8, color=colors[1])
plt.bar(r3, IDM['Test Accuracy'], width=bar_width, label='Improved Distribution Matching (Proposed)', alpha=0.8, color=colors[2])
plt.bar(r4, MI['Test Accuracy'], width=bar_width, label='Model Inversion (Proposed)', alpha=0.8, color=colors[3])

plt.xlabel('IPC', fontsize=14)
plt.ylabel('Test Accuracy', fontsize=14)
# plt.title('Comparison of Test Accuracy Across Benchmarks', fontsize=16)
plt.xticks([r + 1.5*bar_width for r in range(len(DM['IPC']))], DM['IPC'], fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)

plt.tight_layout()
plt.savefig("Test_Accuracy_Comparison.png")
plt.show()
```
