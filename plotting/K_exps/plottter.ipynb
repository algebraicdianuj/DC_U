```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Assuming you have three lists X, Y, and Z
X = [1, 10, 1, 10]    
Y = [45, 45, 450, 450]  
mu_Z_ut = [perc1_k45_mu["unlearning_time"], perc10_k45_mu["unlearning_time"],perc1_k450_mu["unlearning_time"],perc1_k450_mu["unlearning_time"]]
r_Z_ut=[perc1_k45_r["unlearning_time"], perc10_k45_r["unlearning_time"],perc1_k450_r["unlearning_time"],perc1_k450_r["unlearning_time"]]


# Convert lists to numpy arrays
X, Y = np.meshgrid(X, Y)
Z1 = np.array(mu_Z_ut)
Z2=np.array(r_Z_ut)
# Create a figure and a 3D subplot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the 3D surface
surf1 = ax.plot_surface(X, Y, Z1, cmap='viridis', label='MU')

# Plot the second 3D surface (Z2)
surf2 = ax.plot_surface(X, Y, Z2, cmap='plasma', label='R')


# Customize the plot with labels and a title
ax.set_xlabel('Forget Percentage')
ax.set_ylabel('K')
ax.set_zlabel('Unlearning time')
ax.set_title('3D Surface Plot')
# ax.legend()
# Set x-axis ticks to [0, 1050]
# ax.set_yticks([0, 1050])

# # Set x-axis tick labels
# ax.set_yticklabels([0, 1050])


# Rotate the plot by 45 degrees
ax.view_init(elev=10, azim=20)



# Add a color bar which maps values to colors
# fig.colorbar(surf1, ax=ax, shrink=0.5, aspect=10, label='Z1')
# fig.colorbar(surf2, ax=ax, shrink=0.5, aspect=10, label='Z2')

# Add a legend to differentiate between the two surfaces
# ax.legend()

# Display the plot
plt.show()
```
