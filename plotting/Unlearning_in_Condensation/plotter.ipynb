```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


file1 = pd.read_csv("modular_unlearning.csv")
file2 = pd.read_csv("recondensation_training.csv")
file3 = pd.read_csv("retraining.csv")
# file4=  pd.read_csv("fisher_forgetting_stats.csv")
# file5=  pd.read_csv("ntk_scrubbing_stats.csv")


file1.columns = ['MIA', 'RA', 'FA', 'TA', 'UT']
file2.columns = ['MIA', 'RA', 'FA', 'TA', 'UT']
file3.columns = ['MIA', 'RA', 'FA', 'TA', 'UT']
# file4.columns = ['MIA', 'RA', 'FA', 'TA', 'UT']
# file5.columns = ['MIA', 'RA', 'FA', 'TA', 'UT']

bar_width = 0.25
indices = np.arange(len(file1.columns))
r1 = [x for x in indices]
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
# r4 = [x + bar_width for x in r3]
# r5 = [x + bar_width for x in r4]


plt.figure(figsize=(10, 6), dpi=150)

bar1 = plt.bar(r1, file1.iloc[0], width=bar_width, color='#1f77b4', label='modular unlearning', alpha=0.8, edgecolor='black')
bar2 = plt.bar(r2, file2.iloc[0], width=bar_width, color='#ff7f0e', label='remeberance retraining', alpha=0.8, edgecolor='black')
bar3 = plt.bar(r3, file3.iloc[0], width=bar_width, color='#2ca02c', label='retraining from scratch', alpha=0.8, edgecolor='black')
# bar4 = plt.bar(r4, file4.iloc[0], width=bar_width, color='#d62728', label='fisher forgetting', alpha=0.8, edgecolor='black')
# bar5 = plt.bar(r5, file5.iloc[0], width=bar_width, color='#9467bd', label='NTK scrubbing', alpha=0.8, edgecolor='black')


plt.xlabel('Metrics', fontsize=16)
plt.ylabel('Values', fontsize=16)
plt.xticks([r + 1.5 * bar_width for r in range(len(file1.columns))], file1.columns, fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)

plt.tight_layout()
plt.savefig("Metrics_Comparison.png")
plt.show()
```
