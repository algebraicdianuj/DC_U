```python
import matplotlib.pyplot as plt
import seaborn as sns

# Your x and y values and labels
x = [97.86, 46.79,84.16]
y = [75.82, 45.14, 69.47]
labels = ['Pretrained', 'DP-Adam-trained', 'Unlearning']
colors = ['red', 'blue', 'green']  # Colors for each marker
# Set the resolution and size of the figure.
plt.figure(dpi=150, figsize=(8, 8))

# Create a scatter plot using Seaborn. 
# The s parameter is used to increase the size of the markers.
sns.scatterplot(x=x, y=y, s=600, hue=labels, palette=colors,)

# Set labels and title for the plot
plt.xlabel('Train Accuracy',fontsize=18)
plt.ylabel('Test Accuracy',fontsize=18)


# Adding a 45-degree dotted line
plt.plot([0, max(x)+10], [0, max(x)+10], linestyle='dotted', color='grey')

# Adding text labels, with adjusted positions for better visibility
for i, txt in enumerate(labels):
    plt.text(x[i], y[i], txt, fontsize=12, ha='right', va='bottom')  # Adjusted vertical alignment

# Enhancing aesthetics with gridlines
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
