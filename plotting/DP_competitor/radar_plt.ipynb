```python
import numpy as np
import matplotlib.pyplot as plt

def create_radar_plot(data, labels, title):
    # Number of variables
    N = len(labels)
    
    # Compute angle for each axis
    angles = [n / float(N) * 2 * np.pi for n in range(N)]
    angles += angles[:1]  # Repeat the first value to close the circle
    
    # Plot setup
    ax = plt.subplot(111, polar=True)
    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)
    
    # Draw one axe per variable
    plt.xticks(angles[:-1], labels,size=10)
    
    # Draw ylabels
    ax.set_rlabel_position(0)
    plt.yticks([25, 50, 75, 100], ["25%", "50%", "75%", "100%"], color="grey", size=10)
    plt.ylim(0, 100)
    
    # Plot data
    for d, label in data:
        d += d[:1]  # Close the plot
        ax.plot(angles, d, label=label)
        ax.fill(angles, d, alpha=0.1)
    
    plt.title(title)
    plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
    plt.show()

# Data to plot
data = [
    ([97.86, 75.82, 72, 43, 93], "Pre-trained"), 
    ([47.87, 47.31, 36, 29, 83], "DP-Trained"),
    ([86.02, 71.96, 46, 27, 86], "MU-Regularized"),
    # Add other data lines as needed
]

labels = ["TrA", "TeA", "M-AUC", "M-A", "M-PPV"]

create_radar_plot(data, labels, "")

```
