```python
import numpy as np
import matplotlib.pyplot as plt

def create_radar_plot(data, labels, title):
    # Number of variables
    N = len(labels)
    
    # Compute angle for each axis
    angles = [n / float(N) * 2 * np.pi for n in range(N)]
    angles += angles[:1]  # Repeat the first value to close the circle
    
    # Plot setup
    ax = plt.subplot(111, polar=True)
    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)
    
    # Draw one axe per variable
    plt.xticks(angles[:-1], labels,size=10)
    
    # Draw ylabels
    ax.set_rlabel_position(0)
    plt.yticks([25, 50, 75, 100], ["25%", "50%", "75%", "100%"], color="grey", size=10)
    plt.ylim(0, 100)
    
    # Plot data
    for d, label in data:
        d += d[:1]  # Close the plot
        ax.plot(angles, d, label=label)
        ax.fill(angles, d, alpha=0.1)
    
    plt.title(title)
    plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
    plt.show()

# Data to plot
data = [
    ([49.18, 99.735, 91.60, 91.63, 23.12/23.24*100], "Cycle-1"), 
    ([49.78, 99.84, 91.66, 91.33, 21.20/23.24*100], "Cycle-2"), 
    ([49.92, 99.42, 90.48, 90.23, 18.59/23.24*100], "Cycle-3"), 
    ([50.83, 99.58, 90.56, 89.07, 16.67/23.24*100], "Cycle-4"), 
    ([50.02, 99.78, 90.40, 90.11, 14.97/23.24*100], "Cycle-5"), 
]

labels = ["MIA", "RA", "FA", "TeA", "UT%"]

create_radar_plot(data, labels, "Retraining")

```
